```python
import torch
import pandas as pd

def generate_data(n_samples):
    # Generate random 3D Gaussian data
    X = torch.randn(n_samples, 3)

    # Normalize each vector to create the target
    Y = X / torch.norm(X, dim=1, keepdim=True)

    return X.numpy(), Y.numpy()


# Number of samples
n_samples = 10000

X, Y = generate_data(n_samples)

# Create DataFrame
df = pd.DataFrame({
    "x1": X[:, 0],
    "x2": X[:, 1],
    "x3": X[:, 2],
    "y1": Y[:, 0],
    "y2": Y[:, 1],
    "y3": Y[:, 2],
})

# Save to CSV
df.to_csv("sphere_dataset.csv", index=False)

print(df.head())
print("Saved sphere_dataset.csv")
```
